<h1>Build a CNN based on Inception-v3 for image classification on
based on the fashion-mnist dataset. Make settings
models to achieve the required accuracy. After that, on the basis of Siamese networks
build a system to search for similar images in the fashion-mnist dataset.
Visualize the obtained t-SNE results.</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, AveragePooling2D, Dense, Dropout, Flatten, GlobalAveragePooling2D, concatenate, BatchNormalization, Activation
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE


# Download dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Train size 50%
train_size = len(x_train)
subset_size = int(train_size * 0.5)

# Random selection of indices
np.random.seed(42)
subset_indices = np.random.choice(train_size, subset_size, replace=False)

# Selection of the subset
x_train = x_train[subset_indices]
y_train = y_train[subset_indices]

print(f"Original dataset size: {train_size}")
print(f"Subset size (50%): {subset_size}")
print(f"New training dataset size: {len(x_train)}")

# Resize images
x_train = np.array([tf.image.resize(img[..., np.newaxis], (75, 75)).numpy() for img in x_train])
x_test = np.array([tf.image.resize(img[..., np.newaxis], (75, 75)).numpy() for img in x_test])

# Normalize pixel values
x_train = x_train / 255.0
x_test = x_test / 255.0

# Convert to 3 channels cos RGB
x_train = np.repeat(x_train, 3, axis=-1)
x_test = np.repeat(x_test, 3, axis=-1)

# One-hot encode the labels
y_train_one_hot = to_categorical(y_train, 10)
y_test_one_hot = to_categorical(y_test, 10)

# Creating a validation set
x_train, x_val, y_train_one_hot, y_val_one_hot = train_test_split(
    x_train, y_train_one_hot, test_size=0.1, random_state=42
)

print("Dataset shapes:")
print(f"Training: {x_train.shape}")
print(f"Validation: {x_val.shape}")
print(f"Test: {x_test.shape}")


# Inception module
def inception_module(x, filters_1x1, filters_3x3_reduce, filters_3x3, filters_5x5_reduce, filters_5x5, filters_pool_proj):
    conv_1x1 = Conv2D(filters_1x1, (1, 1), padding='same', activation='relu')(x)

    conv_3x3_reduce = Conv2D(filters_3x3_reduce, (1, 1), padding='same', activation='relu')(x)
    conv_3x3 = Conv2D(filters_3x3, (3, 3), padding='same', activation='relu')(conv_3x3_reduce)

    conv_5x5_reduce = Conv2D(filters_5x5_reduce, (1, 1), padding='same', activation='relu')(x)
    conv_5x5 = Conv2D(filters_5x5, (5, 5), padding='same', activation='relu')(conv_5x5_reduce) # uses a lot of RAM, should be used 3x3

    pool_proj = MaxPooling2D((3, 3), strides=(1, 1), padding='same')(x)
    pool_proj = Conv2D(filters_pool_proj, (1, 1), padding='same', activation='relu')(pool_proj)

    output = concatenate([conv_1x1, conv_3x3, conv_5x5, pool_proj], axis=-1)

    return output

# Inception-v3 model
def build_inception_model(input_shape=(75, 75, 3), num_classes=10):
    inputs = Input(shape=input_shape)

    # Initial layers
    x = Conv2D(32, (3, 3), strides=(2, 2), padding='valid', activation='relu')(inputs)
    x = Conv2D(32, (3, 3), padding='valid', activation='relu')(x)
    x = Conv2D(64, (3, 3), padding='same', activation='relu')(x)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    # Inception modules
    x = inception_module(x, filters_1x1=64, filters_3x3_reduce=48, filters_3x3=64,
                     filters_5x5_reduce=16, filters_5x5=32, filters_pool_proj=32)

    x = inception_module(x, filters_1x1=128, filters_3x3_reduce=64, filters_3x3=128,
                     filters_5x5_reduce=32, filters_5x5=64, filters_pool_proj=64)

    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = inception_module(x, filters_1x1=192, filters_3x3_reduce=96, filters_3x3=192,
                     filters_5x5_reduce=32, filters_5x5=96, filters_pool_proj=64)

    # Classification layers
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.4)(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    # Create and compile model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model


# Build and train the Inception-v3 based model
inception_model = build_inception_model()
inception_model.summary()

# Define callbacks for training
checkpoint = ModelCheckpoint('inception_fashion_mnist_best_50pct.h5', monitor='val_accuracy',
                             save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)

# Train the model
history = inception_model.fit(
    x_train, y_train_one_hot,
    validation_data=(x_val, y_val_one_hot),
    epochs=10,
    batch_size=64,
    callbacks=[checkpoint, early_stopping, reduce_lr]
)

# Evaluate the model on test data
test_loss, test_accuracy = inception_model.evaluate(x_test, y_test_one_hot)
print(f"Test accuracy: {test_accuracy:.4f}")

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.savefig('inception_training_history_50pct.png')
plt.show()


# Create Siamese Network for similarity search
def create_embedding_model(base_model):
    """
    Create an embedding model from the trained Inception model
    by removing the classification head
    """
    embedding_model = Model(
        inputs=base_model.input,
        outputs=base_model.layers[-3].output
    )
    return embedding_model


# Create the embedding model
embedding_model = create_embedding_model(inception_model)

def create_siamese_model(embedding_model):
    """
    Create a Siamese network with shared weights
    """
    # Define the inputs for both branches
    input_a = Input(shape=(75, 75, 3))
    input_b = Input(shape=(75, 75, 3))

    # Get embeddings for both inputs using the same embedding model
    embedding_a = embedding_model(input_a)
    embedding_b = embedding_model(input_b)

    # Euclidean distance
    def euclidean_distance(vects):
        x, y = vects
        return tf.sqrt(tf.reduce_sum(tf.square(x - y), axis=1, keepdims=True))

    # Calculate the distance between the two embeddings
    distance = tf.keras.layers.Lambda(euclidean_distance)([embedding_a, embedding_b])

    # Create the Siamese model
    siamese_model = Model(inputs=[input_a, input_b], outputs=distance)

    return siamese_model

# Call the Siamese model
siamese_model = create_siamese_model(embedding_model)


# Generate embeddings for all test images
test_embeddings = embedding_model.predict(x_test)

print(f"Generated embeddings shape: {test_embeddings.shape}")



# Reduce dimensionality with t-SNE
tsne = TSNE(n_components=2, perplexity=30, n_iter=300, random_state=42)
tsne_results = tsne.fit_transform(test_embeddings)

# Plot t-SNE results
plt.figure(figsize=(12, 10))
colors = ['r', 'g', 'b', 'm', 'y', 'c', 'k', 'orange', 'purple', 'brown']
fashion_mnist_labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
                       'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

for i in range(10):
    idxs = np.where(y_test == i)[0]
    plt.scatter(tsne_results[idxs, 0], tsne_results[idxs, 1],
                c=colors[i], label=fashion_mnist_labels[i], alpha=0.6)

plt.title('t-SNE Visualization of Fashion-MNIST Embeddings (50% Training Data)')
plt.legend()
plt.savefig('tsne_fashion_mnist_50pct.png')
plt.show()


# Find similar images

def find_similar_images(query_idx, embeddings, labels, n=5):
    """
    Find n most similar images to the query image based on embedding distance
    """
    query_embedding = embeddings[query_idx].reshape(1, -1)

    # Calculate Euclidean distances
    distances = np.sqrt(np.sum((embeddings - query_embedding) ** 2, axis=1))

    # Get indices of the most similar images (excluding the query itself)
    similar_indices = np.argsort(distances)[1:n+1]

    return similar_indices, distances[similar_indices]


# Visualize some results

def plot_similar_images(query_idx, similar_indices, distances):
    """
    Plot the query image and its similar images
    """
    plt.figure(figsize=(12, 4))

    # Display query image
    plt.subplot(1, 6, 1)
    plt.imshow(x_test[query_idx, :, :, 0], cmap='gray')
    plt.title(f"Query\n{fashion_mnist_labels[y_test[query_idx]]}")
    plt.axis('off')

    # Display similar images
    for i, (idx, dist) in enumerate(zip(similar_indices, distances)):
        plt.subplot(1, 6, i + 2)
        plt.imshow(x_test[idx, :, :, 0], cmap='gray')
        plt.title(f"{fashion_mnist_labels[y_test[idx]]}\nDist: {dist:.2f}")
        plt.axis('off')

    plt.tight_layout()
    plt.savefig(f'similar_images_query_{query_idx}_50pct.png')
    plt.show()

# Visualize similar images for a few random queries
np.random.seed(42)
for _ in range(3):
    query_idx = np.random.randint(0, len(x_test))
    similar_indices, distances = find_similar_images(query_idx, test_embeddings, y_test)
    plot_similar_images(query_idx, similar_indices, distances)
```
